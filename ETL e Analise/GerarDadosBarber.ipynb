```python
from faker import Faker
import pandas as pd
import random
from datetime import datetime
import numpy as np
```

#Gerar Dados


```python
fake = Faker('pt_BR')

NUM_CLIENTES = 3000
NUM_FUNCIONARIOS = 40
NUM_ATENDIMENTOS = 20000

cidades = [
    "Brasília",
    "Taguatinga",
    "Ceilândia",
    "Águas Claras",
    "Sobradinho",
    "Gama"
]

unidades = [
    "Asa Sul",
    "Asa Norte",
    "Taguatinga",
    "Ceilândia"
]

servicos = {
    "Corte": (40, 70),
    "Barba": (25, 45),
    "Pigmentação": (50, 90),
    "Sobrancelha": (20, 35),
    "Hidratação": (60, 120)
}

cargos = [
    "Barbeiro",
    "Gerente",
    "Recepcionista"
]

#Gerar clientes
clientes = []

for i in range(1, NUM_CLIENTES + 1):

    cliente = {
        "id_cliente": i,
        "nome": fake.name(),
        "cidade": random.choice(cidades),
        "idade": random.randint(18, 60),
        "data_cadastro": fake.date_between(
            start_date='-3y',
            end_date='today'
        )
    }

    clientes.append(cliente)

df_clientes = pd.DataFrame(clientes)
```

```python
# INSERIR PROBLEMAS NOS DADOS (INTENCIONAL)
# nomes vazios
for _ in range(20):
    idx = random.randint(0, len(df_clientes)-1)
    df_clientes.loc[idx, "nome"] = None

# clientes duplicados
duplicados = df_clientes.sample(10)
df_clientes = pd.concat([df_clientes, duplicados])
```

```python
# GERAR FUNCIONÁRIOS
funcionarios = []

for i in range(1, NUM_FUNCIONARIOS + 1):

    funcionario = {
        "id_funcionario": i,
        "nome": fake.name(),
        "unidade": random.choice(unidades),
        "cargo": random.choice(cargos)
    }

    funcionarios.append(funcionario)

df_funcionarios = pd.DataFrame(funcionarios)
```

```python
#GERAR ATENDIMENTOS
atendimentos = []

for i in range(1, NUM_ATENDIMENTOS +1):

    servico = random.choice(list(servicos.keys()))

    valor_min, valor_max = servicos[servico]

    valor = round(random.uniform(valor_min, valor_max), 2)

    atendimento = {
        "id_atendimento": i,

        "id_cliente": random.randint(1, NUM_CLIENTES),

        "id_funcionario": random.randint(1, NUM_FUNCIONARIOS),

        "unidade": random.choice(unidades),

        "servico": servico,

        "valor": valor,

        "data": fake.date_between(
            start_date='-2y',
            end_date='today'
        ),

        "forma_pagamento": random.choice([
            "Pix",
            "Cartão",
            "Dinheiro"
        ])
    }

    atendimentos.append(atendimento)

df_atendimentos = pd.DataFrame(atendimentos)

# INSERIR PROBLEMAS NOS DADOS
# valores nulos
for _ in range(50):
    idx = random.randint(0, len(df_atendimentos)-1)
    df_atendimentos.loc[idx, "valor"] = None

# valores negativos
for _ in range(15):
    idx = random.randint(0, len(df_atendimentos)-1)
    df_atendimentos.loc[idx, "valor"] = -50

# serviços escritos errados
erros_servico = {
    "Corte": "cort",
    "Barba": "barb",
    "Pigmentação": "pigmentaçao"
}

for _ in range(30):
    idx = random.randint(0, len(df_atendimentos)-1)

    servico_original = df_atendimentos.loc[idx, "servico"]

    if servico_original in erros_servico:
        df_atendimentos.loc[idx, "servico"] = erros_servico[servico_original]

# duplicados
duplicados = df_atendimentos.sample(25)
df_atendimentos = pd.concat([df_atendimentos, duplicados])
```

```python
# GERAR METAS
metas = []

anos = [2024, 2025]

for ano in anos:
    for mes in range(1, 13):
        for unidade in unidades:

            meta = {
                "ano": ano,
                "mes": mes,
                "unidade": unidade,
                "meta_mensal": random.randint(20000, 50000)
            }

            metas.append(meta)
df_metas = pd.DataFrame(metas)
```

###Exportar CSV

```python
df_clientes.to_csv("Clientes.csv", index=False)

df_funcionarios.to_csv("funcionarios.csv", index=False)

df_atendimentos.to_csv("atendimentos.csv", index=False)

df_metas.to_csv("metas.csv", index=False)


print("\n===== DADOS GERADOS COM SUCESSO =====\n")

print(f"Clientes: {len(df_clientes)}")
print(f"Funcionários: {len(df_funcionarios)}")
print(f"Atendimentos: {len(df_atendimentos)}")
print(f"Metas: {len(df_metas)}")

print("\nArquivos CSV criados!")

```
