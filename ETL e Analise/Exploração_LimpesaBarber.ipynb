#Exploração e Limpeza dos Dados

###Bibliotecas e Importação de Dados

```python
import pandas as pd
import os

clientes = pd.read_csv("/content/Clientes.csv")

funcionarios = pd.read_csv("/content/funcionarios.csv")

atendimentos = pd.read_csv("/content/atendimentos.csv")

metas = pd.read_csv("/content/metas.csv")
```

###Vizualização Inicial

```python
print("\n===== CLIENTES =====")
print(clientes.head())

print("\n===== FUNCIONÁRIOS =====")
print(funcionarios.head())

print("\n===== ATENDIMENTOS =====")
print(atendimentos.head())

print("\n===== METAS =====")
print(metas.head())
```

```python
# ESTRUTURA DOS DADOS

print("ESTRUTURA DOS DATAFRAMES")
print("=====================================")

print("\n===== CLIENTES =====")
print(clientes.info())

print("\n===== FUNCIONÁRIOS =====")
print(funcionarios.info())

print("\n===== ATENDIMENTOS =====")
print(atendimentos.info())

print("\n===== METAS =====")
print(metas.info())
```

```python
# ESTATÍSTICAS DESCRITIVAS
print("ESTATÍSTICAS DESCRITIVAS")
print("=====================================")

print("\n===== CLIENTES =====")
print(clientes.describe())

print("\n===== ATENDIMENTOS =====")
print(atendimentos.describe())
```

```python
#ANÁLISE DE VALORES NULOS

print("VALORES NULOS")
print("=====================================")

print("\n===== CLIENTES =====")
print(clientes.isnull().sum())

print("\n===== FUNCIONÁRIOS =====")
print(funcionarios.isnull().sum())

print("\n===== ATENDIMENTOS =====")
print(atendimentos.isnull().sum())

print("\n===== METAS =====")
print(metas.isnull().sum())
```

```python
# ANÁLISE DE DUPLICADOS
print("DADOS DUPLICADOS")
print("=====================================")

print("Clientes duplicados:")
print(clientes.duplicated().sum())

print("Funcionários duplicados:")
print(funcionarios.duplicated().sum())

print("Atendimentos duplicados:")
print(atendimentos.duplicated().sum())

print("Metas duplicadas:")
print(metas.duplicated().sum())
```

```python
# ANÁLISE DE CATEGORIAS

print("ANÁLISE DE CATEGORIAS")
print("=====================================")

print("\n===== SERVIÇOS =====")
print(atendimentos["servico"].value_counts())

print("\n===== FORMAS DE PAGAMENTO =====")
print(atendimentos["forma_pagamento"].value_counts())

print("\n===== UNIDADES =====")
print(atendimentos["unidade"].value_counts())
```

```python
# VERIFICAÇÃO DE VALORES NEGATIVOS

print("VALORES NEGATIVOS")
print("=====================================")

valores_negativos = atendimentos[
    atendimentos["valor"] < 0
]

print(valores_negativos)
```

```python
# LIMPEZA - CLIENTES

# remover duplicados
clientes = clientes.drop_duplicates()

# remover nomes vazios
clientes = clientes.dropna(subset=["nome"])

# converter data
clientes["data_cadastro"] = pd.to_datetime(
    clientes["data_cadastro"]
)
```

```python
# LIMPEZA - ATENDIMENTOS

#remover duplicados
atendimentos = atendimentos.drop_duplicates()

#remover nulos
atendimentos = atendimentos.dropna(subset=["valor"])

# remover valores negativos
atendimentos = atendimentos[atendimentos["valor"] > 0]

# corrigir serviços escritos errados
atendimentos["servico"] = atendimentos["servico"].replace({
    "cort": "Corte",
    "barb": "Barba",
    "pigmentaçao": "Pigmentação"
})

# converter data
atendimentos["data"] = pd.to_datetime(atendimentos["data"])


```

```python
# VERIFICAÇÃO LIMPEZA

print("\n===== CLIENTES =====")
print(clientes.isnull().sum())

print("\nDuplicados:")
print(clientes.duplicated().sum())

print("\n===== ATENDIMENTOS =====")
print(atendimentos.isnull().sum())

print("\nDuplicados:")
print(atendimentos.duplicated().sum())
```

# CRIAR PASTA DE DADOS TRATADOS

```python
os.makedirs("dados_tratados", exist_ok=True)
```

```python
clientes.to_csv("dados_tratados/clientes_limpos.csv",index=False)

funcionarios.to_csv("dados_tratados/funcionarios_limpos.csv",index=False)

atendimentos.to_csv("dados_tratados/atendimentos_limpos.csv",index=False)

metas.to_csv("dados_tratados/metas_limpos.csv",index=False)
```

```python
# RESUMO FINAL

print("LIMPEZA FINALIZADA")
print("=====================================")

print(f"\nClientes finais: {len(clientes)}")

print(f"Funcionários finais: {len(funcionarios)}")

print(f"Atendimentos finais: {len(atendimentos)}")

print(f"Metas finais: {len(metas)}")

print("\nArquivos tratados exportados!")

print("\nLocal:")
print("dados_tratados/")
```
